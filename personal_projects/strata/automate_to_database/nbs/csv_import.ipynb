```python
import os 
import numpy as np 
import pandas as pd
import psycopg2
```

```python
df = pd.read_csv('../data/onlinefoods.csv')
df.head()
```

```python
file = "onlinefoods"

clean_tbl_name = file.lower().replace(' ', '_').replace('?', '').replace('-', '_').replace(r"/", "_").replace('%', "").replace(')', "").replace(r'(', '').replace('$', "")
clean_tbl_name
```

```python
df.columns = [x.lower().replace(' ', '_').replace('?', '').replace('-', '_').replace(r"/", "_")\
              .replace('%', "").replace(')', "").replace(r'(', '').replace('$', "") for x in df.columns]
df.columns
```

```python
# replacing pandas datatypes with database data types
replacements = {
    'object':'varchar',
    'float64':'float',
    'int64':'int',
    'datetime64':'timestamp',
    'timedelta64[ns]':'varchar'
}
replacements
```

```python
col_str = ", ".join("{} {}".format(n, d) for (n, d) in zip(df.columns, df.dtypes.replace(replacements)))
col_str
```
